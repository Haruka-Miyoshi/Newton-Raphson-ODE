```python
import os
SAVEFILEDIR='./figs'
if not os.path.exists(SAVEFILEDIR):
    os.mkdir(SAVEFILEDIR)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

def f(x):
    return x**2

x=np.linspace(-10, 10, 100)
y=f(x)
plt.title('$y=x^2$')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y)
plt.savefig('./figs/model.png')
plt.show()
```

```python
from model import NewtonRaphson

newton=NewtonRaphson(func=f)
xd=newton.optimization()

yd=f(np.array(xd))
plt.title('$y=x^2$')
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y,c='r', label='$y=x^2$')
plt.scatter(xd,yd,c='b', label='parameter inference')
plt.legend()
plt.show()
```
